## #Reading the training file 

```python
train_data = pd.read_csv(r"C:\Users\home\Desktop\Skillenza\training.csv")
```

```python
train_data.head()
```

## #Checking null counts 

```python
train_data.isnull().sum()
```

```python
train_data["category"].value_counts() #Checking value counts of category column
```

## #Sampling  

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train_data["text"] ,
                                                       train_data["category"],
                                                       test_size = .2,
                                                       random_state = 555)
```

```python
x_train.shape
```

# #Count vectorizer and TDM for train data

```python
from sklearn.feature_extraction.text import CountVectorizer
#CountVectorizer will give the count of each and every word

CV  = CountVectorizer()
trainx = CV.fit_transform(x_train)  #count of each and every unique word
```

```python
print(trainx.shape)
print(trainx)
```

```python
testx= CV.transform(x_test)
print(testx)
```

## #model bulding 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

## #Fit the model

```python
dtree.fit(trainx , y_train)
```

## #Prediction 

```python
pred_value= dtree.predict(testx)
```

## #Confusion Matrix 

```python
from sklearn.metrics import confusion_matrix
tab1 = confusion_matrix(pred_value , y_test)
tab1
```

## #Accuracy 

```python
tab1.diagonal().sum() * 100 / tab1.sum()
```
